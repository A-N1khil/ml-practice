```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Compute X and Y coordinates for points on a sine curve
x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)
print(x)
print(y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
z = np.cos(x)

# Plot
plt.plot(x, y)
plt.plot(x, z)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(["sine", "cosine"])
plt.show()
```
